### Import Modules & Data

```python
import pandas as pd

pd.options.display.max_colwidth = 1000
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', None)
```

```python
raw_data = pd.read_csv('data/CHG-CEE-20_26 August 2021_09.44.csv')

raw_data2 = pd.read_csv('data/CHG-CEE-20_26 August 2021_13.18 3.csv')
```

### Clean Dataframe

#### Remove Header Rows

```python
no_null_df = raw_data.dropna(axis=1, how='all').drop([0,1,2])

# list(no_null_df.columns)
```

```python
# no_null_df['CUMALANG_4_TEXT'].value_counts()
```

#### Make Dummy Variables for Inclusion Questions

```python
no_null_df['inclusion'] = raw_data2['CUINCLUS']

dummied_df = pd.concat([no_null_df, no_null_df.inclusion.str.get_dummies(sep=',')], axis=1)
dummied_df = dummied_df.rename(columns={'1':'Home', '2':'School', '3':'Work', '14':'Shop', '4':'Police',
                      '5':'Other towns or cities', '6':'Health Centres', '7':'Hospital',
                      '8':'Government Departments', '9':'Other Aboriginal Communities',
                      '10':'Social Places', '11':'Shire Transport', '15':'Transport',
                      '12':'Other', '13':'None'})

dummied_df = pd.concat([dummied_df, dummied_df.CUMALANG.str.get_dummies(sep=',')], axis=1)
```

#### Consolidate Time Since Last Paid Work Columns

```python
work_list = []

work_df = dummied_df[['Q20', 'Q20_5_TEXT', 'Q20_8_TEXT']]
for i, row in work_df.iterrows():
    if pd.isna(row['Q20']):
        work_list.append(None)
    if row['Q20'] == 'Number of months ago':
        work_list.append(float(row['Q20_8_TEXT']))
    if row['Q20'] == 'Number of years ago':
        months = int(row['Q20_5_TEXT']) * 12
        work_list.append(months)
    if row['Q20'] == 'I have never worked':
        work_list.append(-1)

dummied_df['last_work'] = work_list
```

#### Filter Out Incomplete/Test Data 

```python
progress = ['45', '100']

filtered_df = dummied_df[dummied_df['Status'] == 'IP Address']
filtered_df = filtered_df[filtered_df['Progress'].isin(progress)]
filtered_df = filtered_df[filtered_df['DE001']!= 'TEST Tim']
```

```python
len(filtered_df)
```

#### Drop Irrelevant Columns

```python
clean_df = filtered_df.drop(columns=['StartDate', 'Status', 'IPAddress', 'Duration (in seconds)', 'Finished',
                                  'RecordedDate', 'ResponseId', 'LocationLatitude', 'LocationLongitude',
                                  'DistributionChannel', 'UserLanguage', 'CONSENT', 'CONSENT2', 'Progress',
                                  'CUMALANG_4_TEXT', 'RecipientLastName', 'RecipientFirstName', 'RecipientEmail',
                                  'ExternalReference', 'WKJBWT_7_TEXT', 'Q76_1', 'Q76_2', 'Q76_3', 'Q20_8_TEXT',
                                  'Q20_5_TEXT', 'Q20', 'CUINCLUS'])
```
